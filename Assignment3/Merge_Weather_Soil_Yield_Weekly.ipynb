```python
import pandas as pd



wheatYieldDf = pd.read_csv('D:\\Grad\\Food_Security\\Assignment3\\KENSAS_winter_wheat_yield_data.csv')
wheatYieldDf = wheatYieldDf[['year', 'state_name', 'county_name', 'yield']]


print(wheatYieldDf)
```

```python
weatherDf = pd.read_csv('D:\\Grad\\Food_Security\\Assignment3\\kansas_weather.csv')

variableDf = weatherDf.drop('date', axis=1)
variableDf = variableDf.drop('state_name', axis=1)
variableDf = variableDf.drop('county_name', axis=1)
variableDf = variableDf.drop('lon', axis=1)
variableDf = variableDf.drop('lat', axis=1)
variableNameList = variableDf.columns

print(weatherDf)
print(variableNameList)
```

```python
soilDf = pd.read_csv('D:\\Grad\\Food_Security\\Assignment3\\state_county_lon_lat_soil.csv')

print(soilDf)
```

```python
soilDf = soilDf.drop('state_name', axis=1)

soil_yield_df = pd.merge(wheatYieldDf, soilDf, on='county_name', how='inner')
print(soil_yield_df)

soil_yield_df.to_csv("D:\\Grad\\Food_Security\\Assignment3\\soil_yield_data.csv")

```

```python
coordinateDf = pd.read_csv("D:\\Grad\\Food_Security\\Assignment3\\state_county_lon_lat.csv")

finalWeatherDf = pd.DataFrame()
weeksNum = 43
variableNum = len(variableNameList)

for index, row in coordinateDf.iterrows():
    
    state_name = row['state_name']
    county_name = row['county_name']
    countyWeatherDf = weatherDf[weatherDf['county_name'] == row['county_name']]
    #print(countyWeatherDf)
    
    
    
    for year in range(2002, 2022):
        countyWeatherDf['date'] = pd.to_datetime(countyWeatherDf['date'], format='%Y%m%d')

        start_date = str(year) + '-09-15'  # The start date of the range
        end_date = str(year + 1) + '-07-15'    # The end date of the range

        mask = (countyWeatherDf['date'] >= start_date) & (countyWeatherDf['date'] <= end_date)
        yearWeatherdf = countyWeatherDf[mask]
        
        # Make sure the weeks number is 43
        row_number = len(yearWeatherdf.reset_index().index)
        while(row_number > weeksNum):
            yearWeatherdf = yearWeatherdf.drop(yearWeatherdf.index[-1])
            row_number = len(yearWeatherdf.reset_index().index)

        # Drop redundant columns
        yearWeatherdf = yearWeatherdf.drop('date', axis=1)
        yearWeatherdf = yearWeatherdf.drop('state_name', axis=1)
        yearWeatherdf = yearWeatherdf.drop('county_name', axis=1)
        yearWeatherdf = yearWeatherdf.drop('lon', axis=1)
        yearWeatherdf = yearWeatherdf.drop('lat', axis=1)
        
        # Put the data which is in the same year in one row
        transposed_data = yearWeatherdf.transpose()
        transposed_data = transposed_data.reset_index(drop=True)
        yearWeatherInOneRow = pd.DataFrame(transposed_data.values.flatten()).T
        
        
        # Add year, state, county information
        yearWeatherInOneRow['year'] = year + 1
        yearWeatherInOneRow['state_name'] = state_name
        yearWeatherInOneRow['county_name'] = county_name

        # Put in the result dataframe
        finalWeatherDf = pd.concat([finalWeatherDf, yearWeatherInOneRow])
        num_columns = len(yearWeatherInOneRow.columns)

        #print(year, num_columns)


print(finalWeatherDf.columns)
# Rename column name
for c in range(weeksNum * variableNum):
    week = c % weeksNum
    
    variableIndex = c // weeksNum
    variableName = variableNameList[variableIndex]
    
    
    finalWeatherDf = finalWeatherDf.rename(columns={finalWeatherDf.columns[c]: 'week' + str(week) + '_' + variableName})

        
print(finalWeatherDf)
finalWeatherDf.to_csv("D:\\Grad\\Food_Security\\Assignment3\\weatherInOneRow.csv", index=False)
    
```

```python
finalWeatherDf = finalWeatherDf.drop('state_name', axis=1)

weather_soil_yield_df = pd.merge(soil_yield_df, finalWeatherDf, on=['county_name', 'year'], how='inner')
print(weather_soil_yield_df)

weather_soil_yield_df.to_csv("D:\\Grad\\Food_Security\\Assignment3\\Weather_Soil_Yield_Data.csv", index=False)
```
